# Naive Bayes Classifier - News categories

```python
import numpy as np
import pandas as pd
import sklearn

training_data = sklearn.datasets.fetch_20newsgroups(subset='train',random_state=42)
test_data = sklearn.datasets.fetch_20newsgroups(subset='test', random_state=42)
```

```python
training_data.data[1][:50]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
vectorizer = TfidfVectorizer()
```

```python
mtx = vectorizer.fit_transform(training_data.data)
y = training_data.target
model.fit(mtx,y)
```

```python
from sklearn.metrics import classification_report

test_mtx = vectorizer.transform(test_data.data)
y_pred = model.predict(test_mtx)
y_test = test_data.target
result = classification_report(y_pred=y_pred,y_true=y_test)
```

```python
print(result)
```
