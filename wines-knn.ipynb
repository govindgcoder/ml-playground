```python
import matplotlib as plt
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import  StandardScaler
from sklearn.metrics import classification_report
from sklearn.pipeline import  Pipeline
```

```python
data = sklearn.datasets.load_wine()
wines = pd.DataFrame(data.data)
classes = pd.DataFrame(data.target)
wines.tail(50)
```

```python
X_train, X_test, y_train, y_test = train_test_split(wines, classes,test_size=0.2, random_state=42)
pipe = Pipeline([('scaler',StandardScaler())])
X_train_values=pipe.fit_transform(X_train)
X_test_values=pipe.transform(X_test)
```

```python
params = {
   'n_neighbors':[i for i in range(100)],
   'weights': ['uniform','distance']
}
model = KNeighborsClassifier()

clfr = GridSearchCV(model, params)

clfr.fit(X_train_values,y_train)
clfr.best_params_
```

```python
model.n_neighbors=75
model.weights='distance'
model.fit(X_train_values,y_train)
y_pred = model.predict(X_test_values)
result = classification_report(y_pred, y_test)
```

```python
print(result)
```
